```python
def create_base_image_white(color255):
    img = np.zeros([400,400,3])

    img[:,:,0] = np.ones([400,400])*color255
    img[:,:,1] = np.ones([400,400])*color255
    img[:,:,2] = np.ones([400,400])*color255
    
    return img
```

```python
def draw_rectangle_plus_color(img, x1, y1, x2, y2, color255):
    start_point = (x1, y1) 
    end_point = (x2, y2) 
    color = (color255, color255, color255) 
    thickness = -1
    image = cv2.rectangle(img, start_point, end_point, color, thickness)
    
    return image
```

```python
def rotate_image(img, angle):
    image_r = ndimage.rotate(img, angle)
    
    return  image_r
```

```python
def cut_and_save(img2, id1, angle, size, angle_id):
    
    img_size_px = int(32 / 2)
    
    w,h = img2.shape[:2]
    
    save1 = img2[int(w / 2) - img_size_px: int(w / 2) + img_size_px, int(h / 2) - img_size_px: int(h / 2) + img_size_px]
    
    cv2.imwrite(dirname + "/img_" + str(id1) + "_" + str(angle) + "_" + str(size) + "_" + str(angle_id) + ".jpg", save1)
```

```python
id1 = 0
angle_id1 = 0

for color255_base in range(1,10):
        
    color255_base_reiz = color255_base * 25

    img1 = create_base_image_white(color255_base_reiz)

    x1 = 200
    y1 = 0
    x2 = 400
    y2 = 400

    for color255_rect in range(1,10):
        
        color255_rect_reiz = color255_rect * 25
        
        img1_rect = draw_rectangle_plus_color(img1, x1, y1, x2, y2, color255_rect_reiz)

        for angle1 in range(0,18):
            
            angle_reiz = angle1 * 10
            
            img1_rect_rot1 = rotate_image(img1_rect, angle_reiz)
            
            size = 0
            
            if (color255_base_reiz == color255_rect_reiz):
                size = 1
            
            cut_and_save(img1_rect_rot1, id1, angle_reiz, size, angle_id1)

            id1 = id1 + 1
            
print(id1)
```

```python
train_images1 = []
file_1_array = []
file_2_array = []

for arr1 in arr:
    id_start = arr1.find("_")
    id_end = arr1.find("_",4)

    angle_start = arr1.find("_",4)
    angle_end = arr1.find("_",angle_start + 1)

    brush_start = arr1.find("_", angle_end)
    brush_end = arr1.find("_", brush_start + 1)

    angle_index_start = arr1.find("_", brush_end)
    angle_index_end = arr1.find(".", angle_index_start + 1)
        
    t_img = cv2.imread(dirname + "\img_" + str(arr1[id_start + 1: id_end]) + "_" + str(arr1[angle_start + 1: angle_end]) + "_" + str(arr1[brush_start + 1: brush_end]) + "_" + str(arr1[angle_index_start + 1: angle_index_end]) + ".jpg",2)

    train_images1.append(t_img)
    
      
    multi_angle_val = 0
    
    if (int(arr1[angle_index_start + 1: angle_index_end]) == 0):
        
        multi_angle_val = int(float(arr1[angle_start + 1: angle_end]) / 10)
    
    if (int(arr1[angle_index_start + 1: angle_index_end]) == 1):
        
        multi_angle_val = int(float(arr1[angle_start + 1: angle_end]) / 10) + 18
    
    if (int(arr1[angle_index_start + 1: angle_index_end]) == 2):
        
        multi_angle_val = int(float(arr1[angle_start + 1: angle_end]) / 10) + 18 * 2
        
    if (int(float(arr1[brush_start + 1: brush_end])) == 1):
        multi_angle_val = 18 * 3
    
    #angle
    file_1_array.append(int(float(multi_angle_val)))
    #brush size
    file_2_array.append(int(float(arr1[brush_start + 1: brush_end])))
    
```

```python
train_images = np.array(train_images1)

train_labels = np.array(file_1_array, dtype=np.uint8)
train_labels_2 = np.array(file_2_array, dtype=np.uint8)
```

```python
train_images = train_images / 255.0
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(32, 32)),
    keras.layers.Dense(128, activation='sigmoid'),
    keras.layers.Dense(55)
])

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])



model.fit(train_images, train_labels, epochs=100)

probability_model = tf.keras.Sequential([model, 
                                         tf.keras.layers.Softmax()])

imgt = train_images[0]

print(imgt.shape)

imgb = (np.expand_dims(imgt,0))

print(imgb.shape)

predictions_single = probability_model.predict(imgb)

print(predictions_single)

np.argmax(predictions_single[0])
```
